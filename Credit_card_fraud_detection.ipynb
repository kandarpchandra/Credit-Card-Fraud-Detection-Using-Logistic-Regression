```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('/content/creditcard_2023.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
credit_card_data.drop('id', axis=1, inplace=True)
```

```python
credit_card_data.head()
```

```python
credit_c_data2 = pd.read_csv("/content/creditcard.csv")
```

```python
credit_c_data2.head()
```

```python
credit_c_data2.drop('Time', axis=1, inplace=True)
```

```python
final_data = pd.concat([credit_card_data, credit_c_data2], axis=0)
```

```python

final_data.head()
```

```python
final_data['Class'].value_counts()
```

```python
legit = final_data[final_data.Class == 0]
fraud = final_data[final_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=362)
```

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
# Function to get user input for features
def get_user_input(feature_names):
    user_features = {}
    print("Please enter the values for each feature:")
    for feature in feature_names:
        while True:
            try:
                value = float(input(f"{feature}: "))
                user_features[feature] = value
                break
            except ValueError:
                print("Invalid input. Please enter a numerical value.")
    return user_features

feature_names = X.columns.tolist()

user_input_features = get_user_input(feature_names)


user_input_df = pd.DataFrame([user_input_features])

prediction = model.predict(user_input_df)
prediction_proba = model.predict_proba(user_input_df)


if prediction[0] == 1:
    print("\n⚠️ WARNING: This transaction is predicted to be FRAUDULENT!")
    print(f"Probability: {prediction_proba[0][1]*100:.2f}%")
else:
    print("\n✅ This transaction is predicted to be LEGITIMATE.")
    print(f"Probability: {prediction_proba[0][0]*100:.2f}%")


print("\nPrediction probabilities:")
print(f"- Legitimate: {prediction_proba[0][0]*100:.2f}%")
print(f"- Fraudulent: {prediction_proba[0][1]*100:.2f}%")
```
